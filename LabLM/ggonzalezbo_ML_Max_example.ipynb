# Learning to sort with a Neural Network (MLPRegressor)

```
import random as rnd
import numpy as np

n=10
Nexamples=1000

X =[]
MAX =[]
for j in range(Nexamples):
    MAX.append(max(np.array(range(n))))
    list1 = list(range(n))
    rnd.shuffle(list1)
    X.append(np.array(list1))
```

```
import pandas as pd
data = pd.DataFrame({'X': [a for a in X], 'MAX': MAX})
data
```

```
from sklearn.neural_network import MLPRegressor
model = MLPRegressor(hidden_layer_sizes=(100,50,20),random_state=1, max_iter=1000, activation='logistic').fit(X,MAX)
MAX_pred = model.predict(X)

import matplotlib.pyplot as plt
plt.plot(model.loss_curve_)
plt.show()
```

```
import pandas as pd
MAX_p = pd.DataFrame({'MAX_pred': MAX_pred})
MAX_p
```
