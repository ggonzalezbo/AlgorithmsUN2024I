## 2. Write in Python a solution that search for the max and second_max sequencialy

```python
%%time
def MaximumPairwiseProductMaxSecondMax(n,L):
  max_v = 0
  second_max = 0
  for i in range(len(L)):
    i=int(i)
    if(L[i] > max_v):
          second_max = max_v
          max_v = L[i]
         # print(max_v , second_max,L)
    elif L[i] > second_max:
        second_max = L[i]
  return  max_v * second_max


a=[2,4,5,6,99,20]
MaximumPairwiseProductMaxSecondMax(2,a)
```
