## Review: Python Lists

- Zero-indexed sequence of arbitrary Python values.
- Slicing syntax: `l[start:stop:step]` copies elements at regular intervals from `start` to `stop`.
- Efficient (`O(1)`) appends and removes from end.
- Comprehension syntax: `[f(x) for x in l if cond(x)]`.

**My own example 19 - Creating a c program that mutiplies 10 times A(600x100) and  B (100x600) matrices**

```python
%%writefile matmul.c
#include <stdio.h>
#include <stdlib.h>
#include <time.h>

void llenarMatriz(int **matriz, int filas, int columnas);
void multiplicarMatrices(int **matrizA, int **matrizB, int **resultado, int filasA, int columnasA);
double promedioMultiplicacion(int **matrizA, int **matrizB, int **resultado, int filas, int columnas, int repeticiones);

int main(void) {
    int filas = 600;
    int columnas = 100;
    unsigned int repeticiones = 10;
    double promedio = 0.0;

    // Asignación dinámica de memoria para las matrices
    int **matrizA = (int **)malloc(filas * sizeof(int *));
    int **matrizB = (int **)malloc(columnas * sizeof(int *));
    int **resultado = (int **)malloc(filas * sizeof(int *));

    for (int i = 0; i < filas; i++) {
        matrizA[i] = (int *)malloc(columnas * sizeof(int));
        matrizB[i] = (int *)malloc(filas * sizeof(int));
        resultado[i] = (int *)malloc(filas * sizeof(int));
    }

    // Llenar las matrices
    llenarMatriz(matrizA, filas, columnas);
    llenarMatriz(matrizB, columnas, filas);

    // Realizar la multiplicación de matrices repetidamente
    promedio = promedioMultiplicacion(matrizA, matrizB, resultado, filas, columnas, repeticiones);
    printf("Promedio en C: %f segundos\n", promedio);

    // Escribir el promedio en un archivo
    FILE *archivoPromedio = fopen("c_average.txt", "w");
    fprintf(archivoPromedio, "%f", promedio);
    fclose(archivoPromedio);

    // Liberar memoria
    for (int i = 0; i < filas; i++) {
        free(matrizA[i]);
        free(matrizB[i]);
        free(resultado[i]);
    }
    free(matrizA);
    free(matrizB);
    free(resultado);

    return 0;
}

void llenarMatriz(int **matriz, int filas, int columnas) {
    for (int i = 0; i < filas; i++) {
        for (int j = 0; j < columnas; j++) {
            matriz[i][j] = rand();
        }
    }
    return;
}

double promedioMultiplicacion(int **matrizA, int **matrizB, int **resultado, int filas, int columnas, int repeticiones) {
    clock_t inicio, fin;
    double promedio = 0.0;
    double ejecucion = 0.0;

    for (int i = 0; i < repeticiones; i++) {
        inicio = clock();
        multiplicarMatrices(matrizA, matrizB, resultado, filas, columnas);
        fin = clock();
        ejecucion = ((double)(fin - inicio)) / CLOCKS_PER_SEC;
        promedio += ejecucion;
        printf("Ejecución %d -> %f segundos\n", i + 1, ejecucion);
    }
    return promedio / ((double)repeticiones);
}

void multiplicarMatrices(int **matrizA, int **matrizB, int **resultado, int filas, int columnas) {
    for (int i = 0; i < filas; i++) {
        for (int j = 0; j < filas; j++) {
            for (int k = 0; k < columnas; k++) {
                resultado[i][j] += matrizA[i][k] * matrizB[k][j];
            }
        }
    }
    return;
}
```

**My own example 22 - Creating a C++ program that mutiplies 10 times A(600x100) and  B (100x600) matrices**

```python
%%writefile matmul.cpp
#include <iostream>
#include <vector>
#include <ctime>
#include <fstream>

using namespace std;

void fillMatrix(vector<vector<int>>& mat, int n, int m) {
    for (int i = 0; i < n; i++) {
        for (int j = 0; j < m; j++) {
            mat[i][j] = rand();
        }
    }
}

void matMul(vector<vector<int>>& matA, vector<vector<int>>& matB, vector<vector<int>>& result, int n, int m) {
    for (int i = 0; i < n; i++) {
        for (int j = 0; j < n; j++) {
            for (int k = 0; k < m; k++) {
                result[i][j] += matA[i][k] * matB[k][j];
            }
        }
    }
}

double multAverage(vector<vector<int>>& matA, vector<vector<int>>& matB, vector<vector<int>>& result, int n, int m, int rep) {
    clock_t start, end;
    double average = 0.0;
    double execution = 0.0;

    for (int i = 0; i < rep; i++) {
        start = clock();
        matMul(matA, matB, result, n, m);
        end = clock();
        execution = static_cast<double>(end - start) / CLOCKS_PER_SEC;
        average += execution;
        cout << "Execution " << i + 1 << " -> " << execution << " seconds" << endl;
    }
    return average / static_cast<double>(rep);
}

int main() {
    int n = 600;
    int m = 100;
    unsigned int rep = 10;
    double average = 0.0;

    // Matrices usando vectores de vectores
    vector<vector<int>> matA(n, vector<int>(m));
    vector<vector<int>> matB(m, vector<int>(n));
    vector<vector<int>> result(n, vector<int>(n));

    // Llenar las matrices
    fillMatrix(matA, n, m);
    fillMatrix(matB, m, n);

    // Realizar multiplicación de matrices repetidas
    average = multAverage(matA, matB, result, n, m, rep);
    cout << "Average in C++: " << average << " seconds" << endl;

    // Escribir el promedio en un archivo
    ofstream fileAvr("c_plus_average.txt");
    fileAvr << average;
    fileAvr.close();

    return 0;
}
```

**My own example 25 - Creating a Java program that mutiplies 10 times A(600x100) and  B (100x600) matrices**

```python
%%writefile MatMul.java
import java.util.Random;
import java.io.FileWriter;
import java.io.IOException;

class MatrixMultiplication {

    public static void main(String[] args) {
        int rowCount = 100; // Número de filas en la matriz A
        int colCount = 10;  // Número de columnas en la matriz B
        int commonDim = 600; // Dimensión común para multiplicación de matrices

        int[][] matrixA = new int[commonDim][rowCount];
        int[][] matrixB = new int[rowCount][colCount];
        int[][] resultMatrix = new int[commonDim][colCount];

        fillMatrix(matrixA, commonDim, rowCount);
        fillMatrix(matrixB, rowCount, colCount);

        double averageTime = calculateAverageExecutionTime(matrixA, matrixB, resultMatrix, commonDim, rowCount, colCount, 10);
        System.out.println("Average Execution Time in Java: " + averageTime + " seconds");

        try {
            FileWriter fileWriter = new FileWriter("java_average.txt");
            fileWriter.write(Double.toString(averageTime));
            fileWriter.close();
        } catch (IOException e) {
            e.printStackTrace();
        }
    }

    static double calculateAverageExecutionTime(int[][] matrixA, int[][] matrixB, int[][] resultMatrix, int commonDim, int rowCount, int colCount, int repetitions) {
        long totalExecutionTime = 0;

        for (int i = 0; i < repetitions; i++) {
            long startTime = System.currentTimeMillis();
            performMatrixMultiplication(matrixA, matrixB, resultMatrix, commonDim, rowCount, colCount);
            long endTime = System.currentTimeMillis();

            long executionTime = endTime - startTime;
            totalExecutionTime += executionTime;
            System.out.println("Execution " + (i + 1) + " -> " + ((double) executionTime / 1000) + " seconds");
        }

        long averageExecutionTime = totalExecutionTime / repetitions;
        double averageTimeInSeconds = (double) averageExecutionTime / 1000;

        return averageTimeInSeconds;
    }

    static void performMatrixMultiplication(int[][] matrixA, int[][] matrixB, int[][] resultMatrix, int commonDim, int rowCount, int colCount) {
        for (int i = 0; i < commonDim; i++) {
            for (int j = 0; j < colCount; j++) {
                for (int k = 0; k < rowCount; k++) {
                    resultMatrix[i][j] += matrixA[i][k] * matrixB[k][j];
                }
            }
        }
    }

    static void fillMatrix(int[][] matrix, int numRows, int numCols) {
        Random random = new Random();

        for (int i = 0; i < numRows; i++) {
            for (int j = 0; j < numCols; j++) {
                matrix[i][j] = random.nextInt(10000);
            }
        }
    }
}
```

```python
%%writefile matMul.js

const fs = require("fs");
const { performance } = require('perf_hooks');

// Función para multiplicar matrices
const multiplyMatrices = (matA, matB, result, n, m) => {
    for (let i = 0; i < n; i++) {
        for (let j = 0; j < n; j++) {
            for (let k = 0; k < m; k++) {
                result[i][j] += matA[i][k] * matB[k][j];
            }
        }
    }
}

// Función para calcular el promedio del tiempo de ejecución
const calculateAverageTime = (matA, matB, result, n, m, rep) => {
    let average = 0;
    let start = 0;
    let end = 0;
    let execution = 0;

    for (let i = 0; i < rep; i++) {
        start = performance.now();
        multiplyMatrices(matA, matB, result, n, m);
        end = performance.now();

        execution = end - start; // en milisegundos
        average += execution;
        console.log(`Ejecución ${i + 1} -> ${execution / 1000} segundos`);
    }

    average /= rep;
    return average / 1000;
}

// Función para generar números aleatorios
const generateRandomNumber = () => {
    const min = 1;
    const max = 100000;
    return Math.floor(Math.random() * (max - min + 1)) + min;
}

// Función principal para multiplicar matrices y calcular el promedio
const matrixMultiplication = () => {
    const n = 600;
    const m = 100;
    const rep = 10;

    // Llenar matrices
    const matA = new Array(n).fill(new Array(m).fill(generateRandomNumber()));
    const matB = new Array(m).fill(new Array(n).fill(generateRandomNumber()));
    const result = new Array(n).fill(new Array(n).fill(0));

    // Repetir la multiplicación de matrices y calcular el promedio del tiempo
    const average = calculateAverageTime(matA, matB, result, n, m, rep);

    console.log(`Promedio en JavaScript: ${average} segundos`);

    // Escribir el promedio en un archivo
    fs.writeFile("js_average.txt", `${average}`, function (err) {
        if (err) {
            console.log("Error al escribir en el archivo:", err);
        }
    });
}

matrixMultiplication();
```

## Why is the Python Version so Much Slower?

```python
# Dynamic typing.
def mul_elemwise(xs, ys):
    return [x * y for x, y in zip(xs, ys)]

mul_elemwise([1, 2, 3, 4], [1, 2 + 0j, 3.0, 'four'])
#[type(x) for x in _]
```

```python
# Interpretation overhead.
source_code = 'a + b * c'
bytecode = compile(source_code, '', 'eval')
import dis; dis.dis(bytecode)
```

## Why is the Python Version so Slow?
- Dynamic typing means that every single operation requires dispatching on the input type.
- Having an interpreter means that every instruction is fetched and dispatched at runtime.
- Other overheads:
  - Arbitrary-size integers.
  - Reference-counted garbage collection.

> This is the paradox that we have to work with when we're doing scientific or numerically-intensive Python. What makes Python fast for development -- this high-level, interpreted, and dynamically-typed aspect of the language -- is exactly what makes it slow for code execution.

- Jake VanderPlas, [*Losing Your Loops: Fast Numerical Computing with NumPy*](https://www.youtube.com/watch?v=EEUXKG97YRw)

```python
def _organic_spread(group):

    if len(group.columns) != 2:
        return pd.Series(index=group.index, data=0.0)

    is_organic = group.columns.get_level_values('Organic').values.astype(bool)
    organics = group.loc[:, is_organic].squeeze()
    non_organics = group.loc[:, ~is_organic].squeeze()
    diff = organics - non_organics
    return diff

def organic_spread_by_region(df):
    """What's the difference between the price of an organic
    and non-organic avocado within each region?
    """
    return (
        df
        .set_index(['Date', 'Region', 'Organic'])
         ['Weighted Avg Price']
        .unstack(level=['Region', 'Organic'])
        .ffill()
        .groupby(level='Region', axis=1)
        .apply(_organic_spread)
    )
```
